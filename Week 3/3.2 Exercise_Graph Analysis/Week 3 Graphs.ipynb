```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
weather = pd.read_csv("C:\\Users\\Gabe\\Documents\\Bellevue University\\Data Mining\\Data\\el_nino_la_nina_1999_2019.csv")
```

```python
# Bar Chart

# Ranking 2

labels = ['2002', '2004', '2006', '2009', '2014', '2015', '2018']
y_pos = np.arange(len(labels))
elNinoStrength = [4,2,2,4,2,8,2]
plt.bar(y_pos, elNinoStrength, align='center', alpha = 0.5)
plt.xticks(y_pos, labels)
plt.xlabel('Year of El Nino Storms')
plt.ylabel('Severity')
plt.title('Severity of El Nino Storms')

plt.show()
```

```python
# Line Chart

# Ranking 1

labels = ['2002', '2004', '2006', '2009', '2014', '2015', '2018']
y_pos = np.arange(len(labels))
elNinoStrength = [4,2,2,4,2,8,2]

plt.plot(y_pos, elNinoStrength)  
plt.xticks(y_pos, labels)
plt.xlabel("Year of El Nino Storms")
plt.ylabel("Severity")
plt.title("Severity of El Nino Storms")

plt.show()
```

```python
# Pie Chart

# Ranking 3

labels = '4', '2', '8'
sizes = [2, 4, 1]
explode = (0, 0.1, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.title("Severity of Occurrences of El Nino Storms")
plt.show()
```
