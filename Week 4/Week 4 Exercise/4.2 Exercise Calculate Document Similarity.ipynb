```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
import string


billboard_df = pd.read_csv("C:\\Users\\Gabe\\Documents\\Bellevue University\\Data Mining\\Week 4\\Week 4 Exercise\\billboard_lyrics_1964_2015.csv", encoding='latin-1')
with open('C:\\Users\\Gabe\\Documents\\Bellevue University\\Data Mining\\Week 4\\Week 4 Exercise\\PositiveWords.txt', 'r') as file:
    positiveWords = file.read().replace('\n', ' ')
with open('C:\\Users\\Gabe\\Documents\\Bellevue University\\Data Mining\\Week 4\\Week 4 Exercise\\NegativeWords.txt', 'r') as file:
    negativeWords = file.read().replace('\n', ' ')

billboard_df['Lyrics'] = billboard_df['Lyrics'].astype(str)
print(billboard_df.head(1))
print(billboard_df.tail(1))
```

```python
corpus = billboard_df['Lyrics']
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print("")
print("vectorized words")
print("")
print(vectorizer.get_feature_names())
print("")
print("Identify Feature Words - Matrix View")
print("")
print( X.toarray())
```

```python
df = pd.DataFrame({'text' : corpus})
#check for positive words and negative words

positive_list = list(positiveWords.split(" "))
negative_list = list(negativeWords.split(" "))

for word in positive_list:
    df['positive'] = df.text.str.count(word)

for w in negative_list:
    df['negative'] = df.text.str.count(w)

df['TotScore'] = df.positive - df.negative

print("")
print(df)

Z = sum(df['TotScore'])
print("")
print("Positive - Positive Value : Negative - Negative Value")
print("")
print("Overall Score:  ",Z)
```

```python
# Extra Credit - Relates to Ch. 5-7

# Jaccard Distance on Two Different Songs

import collections

def intersection(s1, s2) :   
    intersect = s1 & s2 ; 
    return intersect;  
    
def jaccard_index(s1, s2) :  
    size_s1 = len(s1);  
    size_s2 = len(s2);  
    intersect = intersection(s1, s2);   
    size_in = len(intersect);  
    jaccard_in = size_in  / (size_s1 + size_s2 - size_in);  
    return jaccard_in;  
  
def jaccard_distance(jaccardIndex)  : 
    jaccard_dist = 1 - jaccardIndex;   
    return jaccard_dist; 

list_of_lyrics = billboard_df['Lyrics'].to_list()

# I Can't Help Myself - Four Tops
first_song = billboard_df['Lyrics'].iloc[1]
first_song_lyric_list = list(first_song.split(" "))

# How Deep Is Your Love - Calvin Harris
last_song = billboard_df['Lyrics'].iloc[-1]
last_song_lyric_list = list(last_song.split(" "))

first_song_words = set(first_song_lyric_list)
last_song_words = set(last_song_lyric_list)

jaccardIndex_Songs = jaccard_index(first_song_words, last_song_words)

print("Jaccard 1965 Song & 2015 Song Distance = ",jaccard_distance(jaccardIndex_Songs)*100, "%")
```
